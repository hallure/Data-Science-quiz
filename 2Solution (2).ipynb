```python
# Quiz 1

quiz = [
    {"question": "What is the deepest ocean trench?", "answer": "Mariana Trench"},
    {"question": "Who painted the Mona Lisa?", "answer": "Leonardo da Vinci"},
    {"question": "What is the atomic number of hydrogen?", "answer": "1"},
    {"question": "Which city hosted the 2016 Summer Olympics?", "answer": "Rio de Janeiro"},
    {"question": "What is the longest river in the world?", "answer": "Nile"}
]

def run_quiz():
    score = 0
    total_questions = len(quiz)
    print("Welcome to the Quiz Game! Answer the following 5 questions.\n")
    for i, q in enumerate(quiz, 1):
        print(f"Question {i}: {q['question']}")
        user_answer = input("Your answer: ").strip().lower()
        if user_answer == q['answer'].lower():
            print("Correct!\n")
            score += 1
        else:
            print(f"Incorrect. The correct answer is: {q['answer']}\n")
    print(f"Quiz complete! Your score: {score}/{total_questions}")
    percentage = (score / total_questions) * 100
    print(f"Percentage: {percentage}%")

try:
    run_quiz()
except KeyboardInterrupt:
    print("\nQuiz terminated by user.")
```

```python
# Quiz 2

def swap_numbers():
    try:
        num1 = input("Enter the first number: ")
        num2 = input("Enter the second number: ")
        num1 = float(num1)
        num2 = float(num2)
        print(f"Before swapping: First number = {num1}, Second number = {num2}")
        num1, num2 = num2, num1
        print(f"After swapping: First number = {num1}, Second number = {num2}")
    except ValueError:
        print("Error: Please enter valid numeric values.")

swap_numbers()
```

```python
# Quiz 3

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def is_armstrong(n):
    num_str = str(n)
    num_digits = len(num_str)
    return n == sum(int(digit) ** num_digits for digit in num_str)

def is_palindrome(n):
    return str(n) == str(n)[::-1]

def check_number():
    try:
        num = input("Enter a number: ")
        num = int(num)
        if num < 0:
            print("Please enter a non-negative integer.")
            return
        results = []
        if is_prime(num):
            results.append("Prime number")
        if is_armstrong(num):
            results.append("Armstrong number")
        if is_palindrome(num):
            results.append("Palindrome number")
        if results:
            print(f"{num} is a {', '.join(results)}.")
        else:
            print(f"{num} is not a Prime, Armstrong, or Palindrome number.")
    except ValueError:
        print("Error: Please enter a valid integer.")

check_number()
```

```python
# Quiz 4

def fibonacci(n):
    fib_sequence = [0, 1] if n >= 2 else [0] if n == 1 else []
    for i in range(2, n):
        fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
    return fib_sequence

def generate_fibonacci():
    try:
        n = input("Enter the number of Fibonacci terms to generate: ")
        n = int(n)
        if n <= 0:
            print("Error: Please enter a positive integer.")
            return
        sequence = fibonacci(n)
        print(f"The first {n} terms of the Fibonacci sequence are: {sequence}")
    except ValueError:
        print("Error: Please enter a valid integer.")

generate_fibonacci()
```
